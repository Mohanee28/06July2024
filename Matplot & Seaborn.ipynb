```python
import pandas as pd
import os
import names
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Customer = pd.read_excel('Sales.xlsx')
Customer
```

```python
plt.plot(Customer["Region"], Customer["Units Sold"]);
```

```python
cust = Customer.groupby('Region')['Units Sold'].sum()
cust
```

```python
plt.plot(cust.index, cust.values);
```

```python
plt.figure(figsize=(10,3))
plt.title('Units Sold by Region')
plt.xlabel("Region")
plt.ylabel("Unit Sold")
plt.plot(cust.index, cust.values);
```

```python
plt.figure(figsize=(10,3))
plt.title('Units Sold by Region')
plt.xlabel("Region")
plt.ylabel("Unit Sold")
plt.ylim(1,40)
plt.plot(cust.index, cust.values);
```

```python
plt.figure(figsize=(10,3))
plt.title('Units Sold by Region')
plt.xlabel("Region")
plt.ylabel("Unit Sold")
plt.yticks(np.arange(0,41,5))
plt.plot(cust.index, cust.values);
```

```python
plt.figure(figsize=(10,3))
plt.title('Units Sold by Region')
plt.xlabel("Region")
plt.ylabel("Unit Sold")
plt.yticks(np.arange(0,41,5))
plt.plot(cust.index, cust.values, "r--+");
```

```python
plt.figure(figsize=(10,3))
plt.title('Units Sold by Region')
plt.xlabel("Region")
plt.ylabel("Unit Sold")
plt.yticks(np.arange(0,41,5))
plt.plot(cust.index, cust.values, "g--d", mec = "r", mfc = "r");
```

```python
plt.figure(figsize=(10,3))
plt.title('Units Sold by Region')
plt.xlabel("Region")
plt.ylabel("Unit Sold")
plt.yticks(np.arange(0,41,5))
plt.plot(cust.index, cust.values, "g--d", mec = "r", mfc = "r")
plt.legend(labels = ["Unit Sold"]);
```

```python
plt.figure(figsize=(10,3))
plt.title("Unit sold by Region")
plt.xlabel("Region")
plt.ylabel("Unit Sold")
plt.yticks(np.arange(0,40,3))
plt.plot(cust.index, cust.values, "g--d", mec = "r", mfc = "r")
plt.legend(["Unit Sold"])
plt.grid(True)
plt.savefig("Unit sold.png", dpi = 600)
```

##Vertical Bar chart

```python
plt.figure(figsize=(10,3))
plt.title("Unit sold by Region")
plt.xlabel("Region")
plt.ylabel("Unit Sold")
plt.yticks(np.arange(0,40,3))
plt.bar(cust.index, cust.values, color = ["r","g","b","c"]);
```

```python
plt.figure(figsize=(10,3))
plt.title("Unit sold by Region")
plt.xlabel("Region")
plt.ylabel("Unit Sold")
plt.yticks(np.arange(0,40,3))
plt.barh(cust.index, cust.values, color = ["r","g","b","c"]);
```
