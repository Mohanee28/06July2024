##Calculate Central tendancy

```python
data['Population in thousands (2017)'].mean()
```

```python
data['Population in thousands (2017)'].median()
```

```python
# all the below values are equal time repeated values.
data['Population in thousands (2017)'].mode()
```

```python
# select any value by index
data['Population in thousands (2017)'].mode()[0]
```

```python
data['Population in thousands (2017)'].value_counts()
```

```python
#max, min
print("Max value of the Column is:{}". format (data['Population in thousands (2017)'].max()))
print("Min value of the Column is:{}". format (data['Population in thousands (2017)'].min()))
```

#Range -> range of a distribution is min value subracted by max

```python
data['Population in thousands (2017)'].max() - data['Population in thousands (2017)'].min()
```

```python
data['Population in thousands (2017)'].count()
```

```python
data['Population in thousands (2017)'].unique()
```

```python
#Standard Deviation
data['Population in thousands (2017)'].std()
```

```python
#Variance
data['Population in thousands (2017)'].var()
```

```python
#how to calculate percentail
twe = np.percentile(data['Population in thousands (2017)'], 25)
fif = np.percentile(data['Population in thousands (2017)'], 50)
sev = np.percentile(data['Population in thousands (2017)'], 75)
print("25%tail is: {}". format (twe))
print("50%tail is: {}". format (fif))
print("75%tail is: {}". format (sev))
```

```python
#Quartile
Q1 = data['Population in thousands (2017)'].quantile(0.25)
Q2 = data['Population in thousands (2017)'].quantile(0.50)
Q3 = data['Population in thousands (2017)'].quantile(0.75)

print(Q1, Q2, Q3)
```

```python
#IQR

IQR = Q3 - Q1
print("IQR value is: {}".format(IQR))
```
