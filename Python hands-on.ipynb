```python
def two_sum_indices(nums, target):
    """
    Finds the indices of two numbers in the list that sum to the target.
    Uses a dictionary (hash map) for O(n) time complexity.
    """
    # Use a dictionary to store 'value: index'
    num_map = {} 
    
    for index, value in enumerate(nums):
        # Calculate the required complement
        complement = target - value
        
        # Check if the complement is already in the map
        if complement in num_map:
            # If yes, we found the pair! Return the current index and the complement's index
            return [num_map[complement], index]
        
        # If not found, add the current value and its index to the map
        num_map[value] = index
        
    return [] # Return empty list if no solution is found

# Example Usage:
numbers = [2, 7, 11, 15]
target_value = 9
result = two_sum_indices(numbers, target_value)

print(f"List: {numbers}, Target: {target_value}")
print(f"Indices: {result}")
```

```python
class parent():
    def __init__ (self, f, m, mn):
        self.father = f
        self.mother = m
        self.money = mn
    def parent_info (self):
        return f"{self.father} and {self.mother} having assert value of Rs{self.money}"
class child():
    def __init__ (self, cn, cage):
        self.child_name = cn
        self.child_age = cage
    def child_info(self):
        return f"{self.child_name} of {self.father} has created document, when he met {self.child_age} age can be obtain {self.money} Rs"
class gchild(parent, child):
    def __init__ (self, f, m, mn, cn, cage):
        parent.__init__ (self, f, m, mn)
        child.__init__ (self, cn, cage)
    def gchild_info(self):
        print("finaly program excicuted")
        print(self.parent_info())
        print(self.child_info())
    
        
```

```python
class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id
        self.status = "Learning"
    def study(self):
        return f"{self.name} (ID: {self.student_id}) is focused on academic studies."
class Employee:
    def __init__(self, job_title, salary):
        self.job_title = job_title
        self.salary = salary
        self.status = "Working"
    def work(self):
        return f"{self.name} is working as a {self.job_title} for a salary of {self.salary}."
class WorkingStudent(Student, Employee):
    def __init__(self, name, student_id, job_title, salary):
        Student.__init__(self, name, student_id)
        Employee.__init__(self, job_title, salary)
        self.dual_role = True

    def get_info(self):
        print(f"--- {self.name}'s Dual Role Status ---")
        print(self.study())
        print(self.work())
        # Accessing an attribute from the first class in the MRO (Student in this case)
        print(f"Current primary status attribute value: {self.status}") 
        print(f"This person earns {self.salary} while studying.")
```

```python
person1 = WorkingStudent(
    name="Arjun Varma", 
    student_id="CS7890", 
    job_title="Software Intern", 
    salary="Rs 30,000")
person1.get_info()
```

```python
def add (aggregate, *args):
    if aggregate == "int":
        result = 0
    elif aggregate == "str":
        result = ""
    else:
        raise ValueError ("Unknown aggregate type. Must be 'int' or 'str'.")
    for i in args:
        result += i
    return result
```

```python
add("int", 10, 20, 30)
```

```python
def create_user_profile(user_id, **details):
    profile = {'id': user_id}
    print(f"\n-----------processing profile for id: {user_id}-----")
    print (f"Recived details (kwargs): {details}")
    for key, value in details.items():
        profile[key] = value
    if key == "city":
        print(f" -> located in: {value}")
    return profile
```

```python
user1_profile = create_user_profile(
    101,
    name = "Lidu",
    age = 16,
    clas = "SSLC",
    mode_of_trasport = "school bus")
```

```python
user2_profile = create_user_profile(
    105,
    name = "Ranjani",
    age = 18,
    clas = "HSC",
    city = "Cbe")
```

```python
class car():
    def __init__(self):
        self._softwareupdate()        
    def drive(self):
        print("Driving mode")
    def _softwareupdate(self):
        print("Software update")
class rcar(car):
    def setmaxspeed(self, speed):
        self._maxspeed = speed
        super().drive()
        
        return f" {self._maxspeed} is the max speed"
```

```python
redcar = rcar()
redcar._maxspeed = 10
redcar.setmaxspeed(320)
```
