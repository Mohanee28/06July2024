```python
# one sample test of mean
```

```python
# Assemption is, it was obsered that the average purchase made by men of the age 18-25 was 10,000. is it still the same?
#Null Hypothesis -----> the mean is 10,000
#alternative hypothesis -----> the mean is not 10,000

#I can check the significant diffrence between the average purchases
```

```python
df
```

```python
df['Age'].unique()
```

```python
df['Age'] = L1.fit_transform(df['Age'])
```

```python
df['Age'].unique()
```

```python
new_data = df.loc[(df['Age'] == 1) & (df['Gender'] == 1)]
```

```python
new_data.shape
```

```python
sample_size = 1000
```

```python
sample = new_data.sample(sample_size, random_state = 0)
sample
```

```python
sample.shape
```

```python
#proposed mean
pro_mean = 10000
```

```python
sample_mean = sample['Purchase'].mean()
print(sample_mean)
```

```python
#onem sample test
from scipy.stats import ttest_1samp
```

```python
t_stat, p_value = ttest_1samp (sample['Purchase'], pro_mean)
```

```python
print(t_stat, p_value)
```

```python
# ---> P_value is less than 5% that is 0.05, so reject the null hypothesis
# ---> therefore, the mean purchase for the men age 18-25 is not 10,000
```

```python
## one sample test proportion
```

```python
# lets suppose, it was observed that the percentage of women of that spend more than 10000 was 35%, is that still same?
```

```python
# Null - hypothesis --> propotion is 35%
# Alternative hypothesis --> Propotion is not 35%
```

```python
data_new = df.loc[(df['Purchase'] > 10000)]
```

```python
data_new.shape
```

```python
data_new['Gender'].value_counts()
```

```python
count = data_new['Gender'].value_counts()[0] #Females count
nobs = len(data_new['Gender']) #total no of Men and women
p0 = 0.35 # Proposed propotion
print(count, nobs)
```

```python
#percentage of male and female
data_new['Gender'].value_counts()/nobs
```

```python
print(count, nobs)
```

```python
#one sample z-test
from statsmodels.stats.proportion import proportions_ztest
```

```python
z_stat, p_value = proportions_ztest(count = count, nobs = nobs, value = p0, alternative = 'two-sided', prop_var = False)
print(count, nobs)
print(z_stat,'\n', p_value)
```

```python
#P-value is less than 0.05reject the null hypothesis
#The propotion of women spend more than 10000, is not 35%
```

```python
## is the average purchase made by men and women of age 18-25 same?
```

```python
##null hypothesis ---> Average spend are equal
##alternative hypothesis ---> Average spend are not equal
```

```python
data_men = df.loc[(df['Age'] == 1) & (df['Gender'] == 1)]
data_women = df.loc[(df['Age'] == 1) & (df['Gender'] == 0)]
```

```python
data_men.shape
```

```python
data_women.shape
```

```python
data_men_sample = data_men.sample(500, random_state = 0)
data_women_sample = data_women.sample(500, random_state = 0)
```

```python
print(data_men_sample.shape, data_women_sample.shape)
```

```python
print(data_men_sample.Purchase.var())
print(data_women_sample.Purchase.var())
```

```python
print(data_men_sample.Purchase.mean())
print(data_women_sample.Purchase.mean())
```

```python
#calculate F Statestics
from scipy.stats import f
F = data_men_sample.Purchase.mean()/data_women_sample.Purchase.mean()
F
```

```python
# calculate freedom of degree
df1 = len(data_men_sample.Purchase) - 1
df2 = len(data_women_sample.Purchase) - 1
print(df1, df2)
```

```python
# p - Value
import scipy
f_stat = scipy.stats.f.cdf(F, df1, df2)
print(f_stat)
```

```python
# the P-value is grater than 0.05, do not reject null hypothesis.
# The average purchase are same
```
