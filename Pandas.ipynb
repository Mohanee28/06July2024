##Merge and Joins

```python
a = pd.DataFrame({"Name": list("ABCD"), "Age":[12,22,33,44]})
a
```

```python
B = pd.DataFrame({"Student Name": list("XYZ"), "Salary": [122, 1234, 2341]})
B
```

```python
a.merge(B,left_on = "Name", right_on = "Student Name", how = "inner")
```

```python
a.merge(B,left_on = "Name", right_on = "Student Name", how = "left")
```

```python
a.merge(B,left_on = "Name", right_on = "Student Name", how = "right")
```

```python
a.merge(B,left_on = "Name", right_on = "Student Name", how = "outer")
```

```python
a.merge(B, left_on = ["Name", "Age"], right_on = ["Student Name", "Salary"], how = "outer") [["Name", "Salary"]]
```

##Melt - unpivot Function

```python
df = pd.DataFrame({"Sales Rep.": list("ABC"), "Jan": [12, 15, 23], "Feb": [15, 23, 27], "Mar": [2, 26, 29]})
df
```

```python
df.melt(id_vars = "Sales Rep.", value_vars = ["Jan", "Feb", "Mar"])
```

```python
df.melt(id_vars = "Sales Rep.", value_vars = ["Jan", "Feb", "Mar"], var_name = "Month", value_name = "Sales")
```
